### Continue training

```python
# Load a model
model = YOLO('path/to/last.pt')  # load a partially trained model

# Resume training
model.train(resume=True)

# By setting resume=True, the train function will continue training from where it left off, 
# using the state stored in the 'path/to/last.pt' file. If the resume argument is omitted or set to False, 
# the train function will start a new training session.

# Remember that checkpoints are saved at the end of every epoch by default, or at fixed interval using the save_period argument, 
# so you must complete at least 1 epoch to resume a training run.
```
