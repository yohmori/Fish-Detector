### Compress

```python
from PIL import Image
import os

# Specify the folder path containing the PNG images
folder_path = "../Fish_Detector/siamese_and_rummy"

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".png"):
        # Open the PNG image
        image_path = os.path.join(folder_path, filename)
        img = Image.open(image_path)

        # Convert the image to RGB mode if it has an RGBA mode
        if img.mode == "RGBA":
            img = img.convert("RGB")

        # Convert the image to JPEG format and reduce the quality
        # Adjust the quality value (0-100) to control the file size and image quality
        jpeg_path = os.path.join(folder_path, filename.replace(".png", ".jpg"))
        img.save(jpeg_path, "JPEG", quality=70)
```

### Rename

```python
import os

# Specify the folder path containing the files
folder_path = "../Fish_Detector/large_imgs"

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    # Get the full path of the file
    file_path = os.path.join(folder_path, filename)
    
    # Check if the item is a file (not a folder)
    if os.path.isfile(file_path):
        if filename != ".DS_Store":
            # Remove the last four characters from the file name        
            new_filename = filename[:-4]
            
            # Rename the file
            new_file_path = os.path.join(folder_path, new_filename)
            os.rename(file_path, new_file_path)
            
            print(f"Renamed {filename} to {new_filename}")
```
